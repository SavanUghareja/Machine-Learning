```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

```python
X, y = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42,n_features=5)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
df_pca = pd.DataFrame(X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['label'] = y
sns.scatterplot(data=df_pca, x='Principal Component 1', y='Principal Component 2', hue='label', palette='tab10')
```
